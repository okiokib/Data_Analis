```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  # Fixed the import for matplotlib
import seaborn as sns

sns.set_context('notebook')
sns.set_style('whitegrid')
sns.set_palette('Blues_r')  

import warnings
warnings.filterwarnings('ignore')
```

```python
df= pd.read_csv('marketing_data.csv')
df.head()
```

```python
df.shape
```

# Pertanyaan 1
* Apakah ada features yang tidak sesuai?

```python
df.info ()
```

```python
for dtype, col in zip(df.dtypes, df.columns):  # Corrected the typo in 'columns'
    if dtype == 'int64' or dtype == 'float64':  # Fixed the condition
        print(col, dtype)
        print(df[col].min(), df[col].max())
    else:
        print(col, dtype)
        print(df[col].unique())
    print()
```

```python
df.columns = df.columns.str.replace(' ','')
```

```python
df.Income
```

```python
df['Income'] = df['Income'].str.replace('$','')
df['Income'] = df['Income'].str.replace(',','')
```

```python
df['Income'] = df['Income'].astype('float')
```

```python
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])
```

```python
df.info()
```

```python
# Kesimpulan : income(float), Dt_Customer (date)
```

# Pertanyaan 2
* Apakah ada missing value atau data duplikat?

```python
df.isna().sum().sort_values(ascending=False)
```

```python
sns.distplot(df['Income'], kde=False, hist=True)
```

```python
df['Income'].plot(kind='box', figsize=(3,4))
```

```python
df['Income'] = df['Income'].fillna(df['Income'].median())
```
